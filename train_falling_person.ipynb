```python
X=data.drop(columns=['label'])
y=data['label']
```

```python
X.head()
```

```python
sns.countplot(x='label', data=data, palette='dark')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
X=minmax.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,shuffle=True,random_state=11)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(max_iter=1000,solver='liblinear')
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
print("The Score of X_train , y_trian",LR.score(X_train,y_train))
print("The Score of X_test , y_test",LR.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix ,f1_score ,precision_score ,classification_report
CM=confusion_matrix(y_test,y_pred)
print(CM)
sns.heatmap(CM,annot=True, fmt="d" ,cmap="YlGnBu")
plt.show()
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='rbf',max_iter=300)
sv.fit(X_train,y_train)
y_pred1=sv.predict(X_test)
print("The Score of X_train , y_trian",sv.score(X_train,y_train))
print("The Score of X_test , y_test",sv.score(X_test,y_test))
```

```python
CM=confusion_matrix(y_test,y_pred1)
print(CM)
sns.heatmap(CM,annot=True, fmt="d" ,cmap="YlGnBu")
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GBCModel = GradientBoostingClassifier(n_estimators=100,max_depth=5,random_state=33) 
GBCModel.fit(X_train, y_train)
y_pred2=GBCModel.predict(X_test)
#Calculating Details
print('GBCModel Train Score is : ' , GBCModel.score(X_train, y_train))
print('GBCModel Test Score is : ' , GBCModel.score(X_test, y_test))
```

```python
CM=confusion_matrix(y_test,y_pred2)
print(CM)
sns.heatmap(CM,annot=True, fmt="d" ,cmap="YlGnBu")
plt.show()
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.naive_bayes import GaussianNB
GNB=GaussianNB()
GNB.fit(X_train,y_train)
y_pred4=GNB.predict(X_test)
print("The Score of X_train , y_trian",GNB.score(X_train,y_train))
print("The Score of X_test , y_test",GNB.score(X_test,y_test))

print(classification_report(y_test,y_pred4))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNC=KNeighborsClassifier()
KNC.fit(X_train,y_train)
y_pred3=KNC.predict(X_test)
print("The Score of X_train , y_trian",KNC.score(X_train,y_train))
print("The Score of X_test , y_test",KNC.score(X_test,y_test))
print(classification_report(y_test,y_pred3))
```

```python
CM=confusion_matrix(y_test,y_pred3,)
print(CM)
sns.heatmap(CM,annot=True, fmt="d" ,cmap="YlGnBu")
plt.show()
```
